```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('Ventas_Tienda.csv')

# Mostrar las primeras filas del DataFrame
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Desempleo', y='Ventas_Semanales', data=df)
plt.title('Ventas Semanales vs Tasa de Desempleo')
plt.xlabel('Tasa de Desempleo')
plt.ylabel('Ventas Semanales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Desempleo', y='Ventas_Semanales', fill=True, cmap="Blues", thresh=0.05)
plt.title('Densidad de Ventas Semanales vs Tasa de Desempleo')
plt.xlabel('Tasa de Desempleo')
plt.ylabel('Ventas Semanales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(df['Desempleo'], df['Ventas_Semanales'], gridsize=50, cmap='Blues')
plt.colorbar(label='Frecuencia')
plt.title('Ventas Semanales vs Tasa de Desempleo')
plt.xlabel('Tasa de Desempleo')
plt.ylabel('Ventas Semanales')
plt.show()
```

```python
sns.jointplot(data=df, x='Desempleo', y='Ventas_Semanales', kind='hex', cmap='Blues')
plt.suptitle('Ventas Semanales vs Tasa de Desempleo', y=1.02)
plt.show()
```

```python
# Crear los rangos de la tasa de desempleo
bins = [0, 5, 7, 9, 11, 15]
labels = ['0-5', '5-7', '7-9', '9-11', '11-15']
df['Desempleo_Rango'] = pd.cut(df['Desempleo'], bins=bins, labels=labels)

# Crear el boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Desempleo_Rango', y='Ventas_Semanales', data=df)
plt.title('Distribución de Ventas Semanales por Rangos de Tasa de Desempleo')
plt.xlabel('Rango de Tasa de Desempleo')
plt.ylabel('Ventas Semanales')
plt.show()
```
